```python
#                        ðŸš€ Frontier Tech Leaders Programme
#                                ASSIGNMENT 1
#-----------------------------------------------------------------------
#                                ANAS ALHARDI
#-----------------------------------------------------------------------
print("-----WELCOME TO PARTY GUEST LIST-----\n")

#1. Guest List Initialization:
#empty dictionary to store guest information
guests = {}

#2. Adding Guests:
#Add the guests to the dictionary

guests["Alice"] = 28, "alice@email.com"
guests["Bob"] = 35, "bob@email.com"
guests["Charlie"] = 30, "charlie@email.com"
print("After adding guests : \n","=" * 50 , "\n",guests,"\n")
print("=" * 50)

#3. Updating the Guest List:

#Add a guest
guests["David"] = 22, "david@email.com"
print("When adding David : \n","=" * 50 , "\n",guests,"\n")
print("=" * 50)

# Remove a guest
del guests["Bob"]
print("When deleting Bob : \n","=" * 50 , "\n",guests,"\n")
print("=" * 50)


#Get the information for a specific guest
def get_guest_info(guest_name):




  if guest_name in guests :
    age = guests[guest_name][0]
    email = guests[guest_name][1]
    return (f"\n{guest_name} (Age: {age}) is coming to the party! Email {email}")
  else:
    return (f"\n{guest_name} is not on the guest list.")

guest_name = input("Enter the name of the guest: ").strip().capitalize()
print(get_guest_info(guest_name))


# age = guests[guest_name][0]
# email = guests[guest_name][1]
# print(guests[guest_name][1])

print("\n\n","=" * 50)

#5. Guest Counter:
#Display the total number of guests
print(f"\n\nThere are {len(guests)} guests on the list.")

print("\n","=" * 50)


#6. Add the new guest to the dictionary
def add_Guest(selected):

  if selected == "Yes" or selected == "Y" :
    guest_name = input("Enter the name of the new guest: ").strip().capitalize()
    guest_age = int(input("Enter the age of the new guest: "))
    guest_email = input("Enter the email address of the new guest: ").strip().capitalize()
    guests[guest_name] = (guest_age, guest_email)

    print("\nThe guest has been added to the list ## Done !\n\n")
    print("After adding guests : \n","=" * 50 , "\n",guests,"\n")

  elif selected == "No" or selected == "N" :
    print("It was chosen not to add a new guest")

  else:
    print("You must choose (Yes/No) (Y/N)")
    add_Guest(input("Do you want to add a new guest? : (Yes/No) (Y/N):    ").strip().capitalize())

selected = input("Do you want to add a new guest? : (Yes/No) (Y/N) :    ").strip().capitalize()

add_Guest(selected)

#====================================================
print("\n\n","Sort the guests by age")
print("=" * 50)
#=====================================================


#Sort the guests by age
sorted_guests = sorted(guests.items(), key=lambda x: x[1][0])

# Print the sorted guests
for guest_name, (age, email) in sorted_guests:
  print(f"{guest_name} (Age: {age}) is coming to the party! Email: {email}")
```
