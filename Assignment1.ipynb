### BMI Calculator and Health Assessmen

```python
#                        ðŸš€ Frontier Tech Leaders Programme
#                                ASSIGNMENT 1
#-----------------------------------------------------------------------
#                                ANAS ALHARDI
#-----------------------------------------------------------------------
print("-----WELCOME TO BMI CALCULATOR PROGRAM-----\n")


def BimCalcultaor(weight: float, height: float):
  if not type(weight) is int or not type(height) is float:
    print("You must to enter a value")
  else:
    if weight <= 0:
      print("Error Values must be greater than zero")
    elif height <= 0:
      print("Error Values must be greater than zero")
    else:
      if weight >= 0 and height >= 0:
        bmi = weight / float(height * height)
      if bmi < 18.5:
        print(f"Your BMI is {bmi:5.2f}. You have a Underweight weight.")
      if bmi >= 18.5 and bmi < 25:
        print(f"Your BMI is {bmi:5.2f}. You have a Normal weight.")
      if bmi >= 25 and bmi < 30:
        print(f"Your BMI is {bmi:5.2f}. You have a Overweight weight.")
      if bmi >= 30:
        print(f"Your BMI is {bmi:5.2f}. You have a Obesity weight.")


weight = int(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))
BimCalcultaor(weight, height)
```
